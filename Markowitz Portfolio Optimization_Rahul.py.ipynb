```python
#Portfolio Optimization code including Target return


import scipy.optimize as opt
def portfolio_volatility(weights, mean_returns, cov_matrix, risk_aversion):
    return risk_aversion * np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))

def get_optimized_weights(expected_returns, covariance_matrix, target_return, risk_aversion):
    num_assets = len(mean_weights)
    args = (mean_weights, cov_matrix, risk_aversion)
    constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1}, # sum of weights = 1
                   {'type': 'eq', 'fun': lambda x: portfolio_volatility(x, mean_weights, cov_matrix, risk_aversion) - target_return}) # target volatility
    bounds = tuple((0, 1) for asset in range(len(stocks)))
    result = opt.minimize(portfolio_volatility, num_assets*[1./num_assets,], args=args, method='SLSQP', bounds=bounds, constraints=constraints)
    return result.x
target_return = 14
risk_aversion = 2.0
optimized_weights = get_optimized_weights(mean_weights, cov_matrix, target_return, risk_aversion)
print("Optimized Portfolio Weights: ", optimized_weights)
print("Portfolio Volatility: ", portfolio_volatility(optimized_weights, mean_returns, cov_matrix, risk_aversion))
```
